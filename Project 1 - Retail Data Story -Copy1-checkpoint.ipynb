```python
# import Python Libraries
# use Python to manipulate the Kaggle data and look for trends

import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

%matplotlib inline

# Load data
retail = "datasets/OnlineRetail.csv"
df=pd.read_csv(retail,encoding="latin")
```

```python
# Look at the head to verify the data
df.head()
```

```python
#Check columns
df.columns
```

```python
# Rename Columns to match Data Story

df.rename(columns ={'InvoiceNo':'Sale', 'StockCode':'Item', 'Description':'Description', 'Quantity':'Quantity', 'InvoiceDate':'SaleDateTime','UnitPrice':'OriginalPrice','CustomerID':'Customer','Country':'Country'}, inplace=True)
```

```python
# Check that names are changed
df.columns
```

```python
df
```

```python
#Check DF shape
df.shape
```

```python
# Drop any rows with empty cells to clean up and reduce the dataframe size
df_noBlank = df.dropna()
```

```python
#Check DF shape
df_noBlank.shape
```

```python
df_noBlank.dtypes
```

```python
#Remove blanks from price column
df_priceAbovezero = df_noBlank[df_noBlank.OriginalPrice > 0]
```

```python
# check
df_priceAbovezero.OriginalPrice <=0
```

```python
# df had 406829 rows, droping any prices < zero and Df is now 406789
```

```python
#Convert SaleDateTime object to date/time
df_priceAbovezero.SaleDateTime = pd.to_datetime(df_priceAbovezero.SaleDateTime, dayfirst=True)
#df['Time stamp'] = pd.to_datetime(df['Time stamp'], dayfirst=True)
```

```python
df_priceAbovezero.dtypes
```

```python
#Example start here
#make a scatterplot of number of items on each invoice, where invoice = one sale
sns.lmplot(x='InvoiceNo', y='Quantity',data=df)
```

```python
#add more details to scatterplot
sns.lmplot(x='Attack', y='Defense', data=df, 
           fit_reg=False, #remove regression line
           hue='Stage')   #color by evolution stage
```

```python
#boxplot showing stats for each characteristic
sns.boxplot(data=df)
```

```python
#create a dataframe containing the stats for each Pokemon
#drop Total, Stage, and Legendary
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
stats_df.head()
```

```python
#boxplot for stats
sns.boxplot(data=stats_df)
```

```python
#change size of plot
plt.subplots(figsize=(13,7))

#violin plot shows the distribution of attack power for each type
sns.violinplot(x='Type 1', y='Attack', data=df)
```

```python
create color palette
#color Hex numbers are from Bulbapedia (https://bulbapedia.bulbagarden.net/wiki/Category:Type_color_templates)
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```python
#violin plot with Pokemon color palette
plt.subplots(figsize=(13,7))
sns.violinplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

```python
#swarm plot
plt.subplots(figsize=(10,6))
sns.swarmplot(x='Type 1', y='Attack', data=df, palette=pkmn_type_colors)
```

```python
#combine a violin plot with a swarm plot
plt.figure(figsize=(13,7))
sns.violinplot(x='Type 1',
               y='Attack',
               data=df,
               inner=None, #removes bars inside violin plot
               palette=pkmn_type_colors)

sns.swarmplot(x='Type 1',
              y='Attack',
              data=df,
              color='k', #make points black
              alpha=0.7) #slightly transparent

plt.title('Attack by Type')
```
